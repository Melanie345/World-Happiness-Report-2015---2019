### Import 2015 data set from Prepared Data

```python
# Import dataframe of 2015 from Prepared Data
df_2015_checked = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'df_2015_checked.csv'))
```

```python
df_2015_checked.head()
```

### Drop from the 2015 set, the columns Unnamed: 0, Region, Standard Error, Dystopia Residual, then rename Family, Happiness Score & Happiness Rank columns, also add in a 'Year' column so we can see the year when these 5 data sets are merged

```python
df_2015_checked = df_2015_checked.drop(columns = ['Unnamed: 0', 'Region', 'Standard Error', 'Dystopia Residual'])
```

```python
#check to see if columns have been dropped
df_2015_checked.head()
```

```python
#Excellent! Now let's rename the Family, Happiness Score & Happiness Rank columns to match the 2018 & 2019 data sets
df_2015_checked.rename(columns = {'Happiness Score': 'Score'}, inplace = True)
df_2015_checked.rename(columns = {'Happiness Rank': 'Overall rank'}, inplace = True)
df_2015_checked.rename(columns = {'Family': 'Social support'}, inplace = True)
df_2015_checked.rename(columns = {'Country': 'Country or region'}, inplace = True)
```

```python
#check to see if columns have been renamed
df_2015_checked.head()
```

```python
#Great! Now lets ADD a 'Year' column to this 2015 dataframe
df_2015_checked['Year'] = 2015
```

```python
#verifying a Year column has been added and that its rows contain '2015'
df_2015_checked.head()
```

```python
#Export to Prepared Data
df_2015_checked.to_csv(os.path.join(path, '02 Data', 'Prepared Data', 'df_2015_ready_to_merge.csv'))
```
